**Importar librerias**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from google.colab import drive
drive.mount('/content/drive')
```

**Cargar datos**

**Descriptivo general**

```python
descriptivo_general  = datos.describe().round(2)
descriptivo_general.to_excel("resultados_descriptivo_general.xlsx")
descriptivo_general
```

**Descriptivo especifico**

```python
datos_variables_Prod = datos.groupby(['Producto']).describe().round(2)
datos_variables_Prod.to_excel("resultados_producto.xlsx")
datos_variables_Prod
```

**Frecuencias**

```python
freq_abs_turno = datos["Turno"].value_counts()
freq_rel_turno = datos["Turno"].value_counts(normalize=True).round(2)*100
tabla_turno = pd.DataFrame({"Frecuencia absoluta": freq_abs_turno, "Frecuencia relativa": freq_rel_turno})
tabla_turno
```

```python
freq_abs_producto = datos["Producto"].value_counts()
freq_rel_producto = datos["Producto"].value_counts(normalize=True).round(2)*100
tabla_producto = pd.DataFrame({"Frecuencia absoluta": freq_abs_producto, "Frecuencia relativa": freq_rel_producto})
tabla_producto
```

```python
freq_abs_fabrica= datos["Fabrica"].value_counts()
freq_rel_fabrica = datos["Fabrica"].value_counts(normalize=True).round(2)*100
tabla_fabrica = pd.DataFrame({"Frecuencia absoluta": freq_abs_fabrica, "Frecuencia relativa": freq_rel_fabrica})
tabla_fabrica
```

```python
freq_abs_turno_fabrica = datos.groupby(['Año', 'Fabrica', 'Producto', 'Turno']).size()
freq_rel_turno_fabrica = (freq_abs_turno_fabrica / len(datos) * 100).round(2)

tabla_turno_fabrica_simple = pd.DataFrame({
    "Frecuencia Absoluta": freq_abs_turno_fabrica,
    "Frecuencia Relativa (%)": freq_rel_turno_fabrica}).reset_index()
tabla_turno_fabrica_simple
```

**Graficas**

```python

#Indicar cuales son las variables
variable = "Fabrica"
subdivision = "Producto"
unidad = "Numero de personas"

#Gráfico de barras comparativo con dos variables
plt.figure(figsize=(15,5))  #Tamaño de imagen
sns.countplot(x=variable, hue= subdivision,data=datos)
plt.xlabel(variable) #Titulo de eje X
plt.ylabel(unidad) #Titulo de eje Y
plt.show() #Mostrar el gráfico
```

```python

#Indicar cuales son las variables
variable_x = "Altura_cm"
variable_y = "Peso_grs"
subdivision1 = "Fabrica"
subdivision2 = "Producto"
variable_size = "Diametro_cm"
nombre_grafica = "Gráfico de Dispersión"

# Gráfico de dispersión de 3 variables (2 numéricas y 2 categorica)
plt.figure(figsize=(15, 8))  # Tamaño
sns.scatterplot(data=datos, x=variable_x, y=variable_y, hue=subdivision1, style=subdivision2, size=variable_size, sizes=(10, 200))  # Generar según variable
plt.title(nombre_grafica)  # título
plt.xlabel(variable_x)  # Eje X
plt.ylabel(variable_y)  # Eje Y
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')  # Ajustar la leyenda
plt.tight_layout()
plt.show()  # mostrar
```

```python
variable_numerica = "Peso_grs"
variable_categorica_agrupacion = "Fabrica"
variable_categorica_agrupacion2 = "Turno"
titulo_grafico = "Boxplot de Salarios Vs Estrato social vs Cargo"

#Gráfico Boxplot a una sola variable númerica según dos categorica
plt.figure(figsize=(12,6))
sns.boxplot(x=variable_categorica_agrupacion, y=variable_numerica, hue=variable_categorica_agrupacion2, data=datos, palette="Set1")
plt.title(titulo_grafico)
plt.show()
```
